## **MUTABLITY**

```python
customer = {
    "id": 1234,
    "name": "John Smith",
    "email": "john@example.com"
}

# Update the customer's email address
customer["email"] = "john.smith@example.com"

# Add a new key-value pair
customer["phone"] = "555-555-1212"

# Remove the customer's ID
del customer["id"]

print(customer)
```
